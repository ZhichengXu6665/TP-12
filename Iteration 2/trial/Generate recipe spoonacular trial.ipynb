```python
import requests

# Replace 'YOUR_API_KEY' with your actual Spoonacular API key
BASE_URL = 'https://api.spoonacular.com'

def self_search_recipes_by_ingredients(ingredients, number_of_recipes=5):
    # Construct the search URL
    search_url = f"{BASE_URL}/recipes/findByIngredients"
    
    # Parameters for the API request
    params = {
        'ingredients': ','.join(ingredients),
        'number': number_of_recipes,
        'apiKey': API_KEY
    }
    
    # Make the API request
    response = requests.get(search_url, params=params)
    
    # Check if the request was successful
    if response.status_code == 200:
        recipes = response.json()
        return recipes
    else:
        print(f"Error: {response.status_code}")
        return None
```

```python
# Example usage
ingredients_list = ['hemp oil cold pressed', 'white toast soft', 'eggs f/range x1 12pk', 'spaghetti rich tomato', '%c 1/2 fam ht chicken']
recipes = self_search_recipes_by_ingredients(ingredients_list)

if recipes:
    for recipe in recipes:
        print(f"Recipe: {recipe['title']}")
        print(f"Used Ingredients: {', '.join([ingredient['name'] for ingredient in recipe['usedIngredients']])}")
        print(f"Missing Ingredients: {', '.join([ingredient['name'] for ingredient in recipe['missedIngredients']])}")
        print(f"URL: https://spoonacular.com/recipes/{recipe['title'].replace(' ', '-')}-{recipe['id']}")
        print()
```

```python
def get_recipe_details(ids, includeNutrition=None):
        """ Get information about a recipe.
            https://spoonacular.com/food-api/docs#get-recipe-information
        """
        endpoint = "recipes/{ids}/information".format(ids=ids)
        url_query = {}
        url_params = {"includeNutrition": includeNutrition}
        return requests.get(endpoint, method="GET", query_=url_query, params_=url_params)
    
```

```python
def classify_a_grocery_product(self, product):
        """ Given a grocery product title, this endpoint allows
            you to detect what basic ingredient it is.
            https://spoonacular.com/food-api/docs#classify-a-grocery-product
        """
        endpoint = "food/products/classify"
        url_json = product
        quota = {'requests': 0, 'tinyrequests': 1, 'results': 0}
        return self._make_request(endpoint, method="POST", json_=url_json)
```
