```python
import requests

# Replace 'YOUR_API_KEY' with your actual Spoonacular API key
API_KEY = '77268583b575469e898efe16e4e81960'
BASE_URL = 'https://api.spoonacular.com'

def search_recipes_by_ingredients(ingredients, number_of_recipes=5):
    # Construct the search URL
    search_url = f"{BASE_URL}/recipes/findByIngredients"
    
    # Parameters for the API request
    params = {
        'ingredients': ','.join(ingredients),
        'number': number_of_recipes,
        'apiKey': API_KEY
    }
    
    # Make the API request to search for recipes
    response = requests.get(search_url, params=params)
    
    if response.status_code == 200:
        recipes = response.json()
        
        # Get detailed information for each recipe
        detailed_recipes = []
        for recipe in recipes:
            recipe_id = recipe['id']
            details = get_recipe_details(recipe_id)
            if details:
                detailed_recipes.append(details)
        
        return detailed_recipes
    else:
        print(f"Error: {response.status_code}")
        return None

def get_recipe_details(recipe_id):
    # Construct the URL for detailed recipe information
    recipe_url = f"{BASE_URL}/recipes/{recipe_id}/information"
    
    # Parameters for the API request
    params = {
        'includeNutrition': 'true',  # Include nutritional information
        'apiKey': API_KEY
    }
    
    # Make the API request for recipe details
    response = requests.get(recipe_url, params=params)
    
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching details for recipe ID {recipe_id}: {response.status_code}")
        return None

def extract_main_nutritional_info(nutrition):
    # Extract main nutritional info: Calories, Carbs, Fat, Protein
    nutrients = nutrition.get('nutrients', [])
    
    main_info = {}
    for nutrient in nutrients:
        name = nutrient['name']
        if name in ['Calories', 'Carbohydrates', 'Fat', 'Protein']:
            main_info[name] = f"{nutrient['amount']} {nutrient['unit']}"
    
    return main_info

# Example usage
ingredients = ['tomato', 'cheese', 'onion']
recipes = search_recipes_by_ingredients(ingredients)

if recipes:
    for recipe in recipes:
        print(f"Title: {recipe['title']}")
        print(f"Ingredients: {', '.join([ingredient['name'] for ingredient in recipe['extendedIngredients']])}")
        print(f"Instructions: {recipe.get('instructions', 'No instructions available')}")

        # Extract and display main nutritional info
        nutrition_info = extract_main_nutritional_info(recipe['nutrition'])
        print("Nutritional Info:")
        for key, value in nutrition_info.items():
            print(f"{key}: {value}")
        
        print("\n")
```
